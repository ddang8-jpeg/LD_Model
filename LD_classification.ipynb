```python
import pandas as pd
from matplotlib.pyplot import subplots
import statsmodels.api as sm
from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
                         summarize)
```

Import data for tower counts and LD outbreaks for each NTA code.

```python
towers = pd.read_csv('data/cleaned_data/nta_tower_count.csv')
LD_cases = pd.read_csv('data/cleaned_data/NYC.csv')
demographics = pd.read_csv('data/cleaned_data/NYC_demographics.csv')
```

```python
brook_cases = LD_cases.loc[lambda LD_cases: LD_cases['NTA code'].str.contains('BK', na=False)]
brook_demo = demographics.loc[lambda demographics: demographics['NTA code'].str.contains('BK', na=False)].drop(columns='NTA name')
```

```python
towers.set_index('NTA code', inplace=True)
brook_cases.set_index('NTA code', inplace=True)
brook_demo.set_index('NTA code', inplace=True)
brook_cases.head(3)
```

```python
brook_demo.head(3)
```

```python
data = brook_demo.join(brook_cases).join(towers)
data.fillna(0);
```

```python
with open('data/cleaned_data/NTA_LD_towers.csv', 'w', newline='') as file:
    data.to_csv(file)
```
